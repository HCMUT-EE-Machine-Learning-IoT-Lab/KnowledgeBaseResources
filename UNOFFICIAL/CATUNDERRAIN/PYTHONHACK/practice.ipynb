Đây là danh sách các hàm dựng sẵn trong Python, mỗi hàm có chức năng riêng biệt. Dưới đây là giải thích ngắn gọn về một số hàm phổ biến cùng với ví dụ minh họa:

1. **abs()**: Trả về giá trị tuyệt đối của một số.
   - Ví dụ: `abs(-5)` → kết quả: `5`.

2. **all()**: Trả về `True` nếu tất cả các phần tử của iterable là `True`.
   - Ví dụ: `all([True, True, False])` → kết quả: `False`.

3. **any()**: Trả về `True` nếu ít nhất một phần tử trong iterable là `True`.
   - Ví dụ: `any([True, False, False])` → kết quả: `True`.

4. **bin()**: Chuyển đổi một số nguyên thành chuỗi nhị phân.
   - Ví dụ: `bin(10)` → kết quả: `'0b1010'`.

5. **bool()**: Chuyển đổi giá trị thành kiểu boolean (`True` hoặc `False`).
   - Ví dụ: `bool(0)` → kết quả: `False`.

6. **callable()**: Kiểm tra xem đối tượng có thể gọi được hay không.
   - Ví dụ: `callable(print)` → kết quả: `True`.

7. **chr()**: Trả về ký tự đại diện cho giá trị ASCII của một số nguyên.
   - Ví dụ: `chr(97)` → kết quả: `'a'`.

8. **dict()**: Tạo một từ điển.
   - Ví dụ: `dict(a=1, b=2)` → kết quả: `{'a': 1, 'b': 2}`.

9. **enumerate()**: Trả về một đối tượng liệt kê (chỉ số và giá trị).
   - Ví dụ: `list(enumerate(['a', 'b', 'c']))` → kết quả: `[(0, 'a'), (1, 'b'), (2, 'c')]`.

10. **eval()**: Thực thi một chuỗi lệnh Python.
    - Ví dụ: `eval('5 + 5')` → kết quả: `10`.

11. **filter()**: Lọc các phần tử của một iterable dựa trên một điều kiện.
    - Ví dụ: `list(filter(lambda x: x > 5, [1, 2, 6, 7]))` → kết quả: `[6, 7]`.

12. **float()**: Chuyển đổi giá trị thành số thực.
    - Ví dụ: `float(5)` → kết quả: `5.0`.

13. **hash()**: Trả về giá trị băm của một đối tượng.
    - Ví dụ: `hash('hello')` → một số nguyên đại diện cho chuỗi `'hello'`.

14. **id()**: Trả về địa chỉ của đối tượng trong bộ nhớ.
    - Ví dụ: `id(5)` → một số nguyên đại diện cho vị trí của số `5` trong bộ nhớ.

15. **len()**: Trả về độ dài của một đối tượng (chuỗi, danh sách, v.v.).
    - Ví dụ: `len('hello')` → kết quả: `5`.

16. **list()**: Chuyển đổi một iterable thành danh sách.
    - Ví dụ: `list('hello')` → kết quả: `['h', 'e', 'l', 'l', 'o']`.

17. **max()**: Trả về giá trị lớn nhất trong một iterable.
    - Ví dụ: `max([1, 2, 3])` → kết quả: `3`.

18. **min()**: Trả về giá trị nhỏ nhất trong một iterable.
    - Ví dụ: `min([1, 2, 3])` → kết quả: `1`.

19. **pow()**: Trả về lũy thừa của hai số.
    - Ví dụ: `pow(2, 3)` → kết quả: `8`.

20. **print()**: In ra màn hình.
    - Ví dụ: `print('Hello, world!')` → kết quả: `Hello, world!`.

21. **range()**: Trả về một đối tượng range từ giá trị bắt đầu đến giá trị kết thúc.
    - Ví dụ: `list(range(5))` → kết quả: `[0, 1, 2, 3, 4]`.

22. **round()**: Làm tròn số đến số chữ số thập phân nhất định.
    - Ví dụ: `round(5.678, 2)` → kết quả: `5.68`.

23. **sum()**: Tính tổng các phần tử trong iterable.
    - Ví dụ: `sum([1, 2, 3])` → kết quả: `6`.

24. **zip()**: Kết hợp các iterable thành một iterator của các bộ giá trị.
    - Ví dụ: `list(zip([1, 2, 3], ['a', 'b', 'c']))` → kết quả: `[(1, 'a'), (2, 'b'), (3, 'c')]`.
