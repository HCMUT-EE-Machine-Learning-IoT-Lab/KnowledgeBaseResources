y l danh sch cc hm dng sn trong Python, mi hm c chc nng ring bit. Di y l gii thch ngn gn v mt s hm ph bin cng vi v d minh ha:

1. **abs()**: Tr v gi tr tuyt i ca mt s.
   - V d: `abs(-5)`  kt qu: `5`.

2. **all()**: Tr v `True` nu tt c cc phn t ca iterable l `True`.
   - V d: `all([True, True, False])`  kt qu: `False`.

3. **any()**: Tr v `True` nu t nht mt phn t trong iterable l `True`.
   - V d: `any([True, False, False])`  kt qu: `True`.

4. **bin()**: Chuyn i mt s nguyn thnh chui nh phn.
   - V d: `bin(10)`  kt qu: `'0b1010'`.

5. **bool()**: Chuyn i gi tr thnh kiu boolean (`True` hoc `False`).
   - V d: `bool(0)`  kt qu: `False`.

6. **callable()**: Kim tra xem i tng c th gi c hay khng.
   - V d: `callable(print)`  kt qu: `True`.

7. **chr()**: Tr v k t i din cho gi tr ASCII ca mt s nguyn.
   - V d: `chr(97)`  kt qu: `'a'`.

8. **dict()**: To mt t in.
   - V d: `dict(a=1, b=2)`  kt qu: `{'a': 1, 'b': 2}`.

9. **enumerate()**: Tr v mt i tng lit k (ch s v gi tr).
   - V d: `list(enumerate(['a', 'b', 'c']))`  kt qu: `[(0, 'a'), (1, 'b'), (2, 'c')]`.

10. **eval()**: Thc thi mt chui lnh Python.
    - V d: `eval('5 + 5')`  kt qu: `10`.

11. **filter()**: Lc cc phn t ca mt iterable da trn mt iu kin.
    - V d: `list(filter(lambda x: x > 5, [1, 2, 6, 7]))`  kt qu: `[6, 7]`.

12. **float()**: Chuyn i gi tr thnh s thc.
    - V d: `float(5)`  kt qu: `5.0`.

13. **hash()**: Tr v gi tr bm ca mt i tng.
    - V d: `hash('hello')`  mt s nguyn i din cho chui `'hello'`.

14. **id()**: Tr v a ch ca i tng trong b nh.
    - V d: `id(5)`  mt s nguyn i din cho v tr ca s `5` trong b nh.

15. **len()**: Tr v  di ca mt i tng (chui, danh sch, v.v.).
    - V d: `len('hello')`  kt qu: `5`.

16. **list()**: Chuyn i mt iterable thnh danh sch.
    - V d: `list('hello')`  kt qu: `['h', 'e', 'l', 'l', 'o']`.

17. **max()**: Tr v gi tr ln nht trong mt iterable.
    - V d: `max([1, 2, 3])`  kt qu: `3`.

18. **min()**: Tr v gi tr nh nht trong mt iterable.
    - V d: `min([1, 2, 3])`  kt qu: `1`.

19. **pow()**: Tr v ly tha ca hai s.
    - V d: `pow(2, 3)`  kt qu: `8`.

20. **print()**: In ra mn hnh.
    - V d: `print('Hello, world!')`  kt qu: `Hello, world!`.

21. **range()**: Tr v mt i tng range t gi tr bt u n gi tr kt thc.
    - V d: `list(range(5))`  kt qu: `[0, 1, 2, 3, 4]`.

22. **round()**: Lm trn s n s ch s thp phn nht nh.
    - V d: `round(5.678, 2)`  kt qu: `5.68`.

23. **sum()**: Tnh tng cc phn t trong iterable.
    - V d: `sum([1, 2, 3])`  kt qu: `6`.

24. **zip()**: Kt hp cc iterable thnh mt iterator ca cc b gi tr.
    - V d: `list(zip([1, 2, 3], ['a', 'b', 'c']))`  kt qu: `[(1, 'a'), (2, 'b'), (3, 'c')]`.
